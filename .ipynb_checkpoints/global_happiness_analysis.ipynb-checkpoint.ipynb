```python
# ---------------------------------------------------------------
# ðŸ”¹ Step 8: Adding Year Columns
# ðŸª„ What to Expect:
# Adding a 'year' column to each dataset for clear time tracking.
#
# âœ… Accomplished:
# âœ… Year columns added to all datasets
# âœ… Temporal tracking prepared
# ---------------------------------------------------------------

df_2016['year'] = 2016
df_2017['year'] = 2017
df_2018['year'] = 2018
df_2019['year'] = 2019
```

```python
# ---------------------------------------------------------------
# ðŸ”¹ Step 10: Cleaning Combined Dataset
# ðŸª„ What to Expect:
# Initial cleaning of `df_all` by dropping irrelevant columns and inspecting data.
#
# âœ… Accomplished:
# âœ… Dataset cleaned to retain relevant columns
# âœ… Basic structure prepared for deeper filtering
# ---------------------------------------------------------------

df_all.dropna(subset=['country', 'happiness_score'], inplace=True)
df_all.head()
```

```python
# ---------------------------------------------------------------
# ðŸ”¹ Step 12: Happiness Score Distribution (Histogram)
# ðŸª„ What to Expect:
# Visualizing the distribution of happiness scores globally.
#
# âœ… Accomplished:
# âœ… Histogram created for happiness scores
# âœ… Visual understanding of score spread achieved
# ---------------------------------------------------------------

plt.figure(figsize=(10, 6))
sns.histplot(df['happiness_score'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Happiness Scores (2015)', fontsize=16)
plt.xlabel('Happiness Score')
plt.ylabel('Frequency')
plt.show()
```

```python
# ---------------------------------------------------------------
# ðŸ”¹ Step 13: Top 10 and Bottom 10 Happiest Countries (Final Clean)
# ---------------------------------------------------------------

# Top 10 happiest countries
top_10 = df.nlargest(10, 'happiness_score')

plt.figure(figsize=(10, 6))
sns.barplot(
    data=top_10,
    x='happiness_score',
    y='country',
    hue='country',
    dodge=False,
    palette='Greens_r',
    legend=False
)
plt.title('Top 10 Happiest Countries (2015)', fontsize=16)
plt.xlabel('Happiness Score')
plt.ylabel('Country')
plt.show()

# Bottom 10 happiest countries
bottom_10 = df.nsmallest(10, 'happiness_score')

plt.figure(figsize=(10, 6))
sns.barplot(
    data=bottom_10,
    x='happiness_score',
    y='country',
    hue='country',
    dodge=False,
    palette='Reds_r',
    legend=False
)
plt.title('Bottom 10 Happiest Countries (2015)', fontsize=16)
plt.xlabel('Happiness Score')
plt.ylabel('Country')
plt.show()
```

```python
# ---------------------------------------------------------------
# ðŸ”¹ Step 14 (Final): Clean Multi-Year Loading with All Years
# ---------------------------------------------------------------

import pandas as pd

years = [2015, 2016, 2017, 2018, 2019]
dfs = []

for year in years:
    file_name = f'{year}.csv'
    df_year = pd.read_csv(file_name)
    df_year['year'] = year

    # Normalize column names BEFORE selection
    df_year.columns = df_year.columns.str.strip().str.lower().str.replace(' ', '_').str.replace(r'[().]', '', regex=True)

    rename_map = {
        'country_or_region': 'country',
        'score': 'happiness_score',
        'happinessscore': 'happiness_score',
        'economygdppercapita': 'economy_gdp_per_capita',
        'social_support': 'family',
        'healthylife_expectancy': 'health_life_expectancy',
        'healthlife_expectancy': 'health_life_expectancy',
        'freedomtomakelifechoices': 'freedom',
        'trustgovernmentcorruption': 'trust_government_corruption',
        'perceptionsofcorruption': 'trust_government_corruption',
    }
    df_year = df_year.rename(columns=rename_map)

    # Filter to required columns if present
    required_columns = ['country', 'happiness_score', 'economy_gdp_per_capita',
                        'family', 'health_life_expectancy', 'freedom',
                        'trust_government_corruption', 'generosity', 'year']

    df_year = df_year[[col for col in required_columns if col in df_year.columns]]

    dfs.append(df_year)

# Concatenate all years
df_all = pd.concat(dfs, ignore_index=True)

# Drop rows missing critical values
df_all = df_all.dropna(subset=['country', 'happiness_score'])

# Export to CSV for your Streamlit dashboard
df_all.to_csv('merged_happiness_data.csv', index=False)

# Confirm structure
print("Shape of merged dataset:", df_all.shape)
print("Available years:", sorted(df_all['year'].unique()))
df_all.head()
```

```python
# ---------------------------------------------------------------
# ðŸ”¹ Step 15A: Column Standardization & Filtering
# ðŸª„ What to Expect:
# Refines df_all with consistent column names,
# filters relevant columns, drops rows with missing data.
# ---------------------------------------------------------------

# Clean and standardize column names
df_all.columns = df_all.columns.str.strip().str.lower().str.replace(' ', '_').str.replace(r'\(.*?\)', '', regex=True)

# Define required columns explicitly
required_columns = [
    'country', 'happiness_score', 'economy_gdp_per_capita',
    'family', 'health_life_expectancy', 'freedom',
    'trust_government_corruption', 'generosity', 'year'
]

# Filter columns if they exist
df_all = df_all[[col for col in required_columns if col in df_all.columns]]

# Drop missing values in critical columns
df_all = df_all.dropna(subset=['country', 'happiness_score'])

# Confirm structure
print("âœ… Step 15A complete. Clean dataset ready.")
print("Shape:", df_all.shape)
print("Available years:", df_all['year'].unique())
df_all.head()
```

```python
# Remove duplicate columns by keeping the first occurrence
df_all = df_all.loc[:, ~df_all.columns.duplicated()]

# Confirm cleanup
print("Columns after removing duplicates:", df_all.columns.tolist())
```

```python
# ---------------------------------------------------------------
# ðŸ”¹ Step 15B: Year-over-Year Analysis for Top 5 Countries
# ðŸª„ What to Expect:
# Tracks and visualizes happiness scores for the top 5 happiest countries in 2015 across 2015â€“2019.
# ---------------------------------------------------------------

# Identify the top 5 countries in 2015
top5_2015 = df_all[df_all['year'] == 2015].nlargest(5, 'happiness_score')['country'].tolist()
print("Top 5 Countries in 2015:", top5_2015)

# Filter dataset for these countries across all years
df_top5 = df_all[df_all['country'].isin(top5_2015)]

# Plot year-over-year happiness trends
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.lineplot(data=df_top5, x='year', y='happiness_score', hue='country', marker='o')
plt.title('Year-over-Year Happiness Trend for Top 5 Countries (2015â€“2019)', fontsize=16)
plt.xlabel('Year')
plt.ylabel('Happiness Score')
plt.legend(title='Country')
plt.show()
```
