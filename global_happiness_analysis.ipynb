```python
# üîπ Step 14 (Final): Clean Multi-Year Loading with All Years
# ---------------------------------------------------------------

import pandas as pd

years = [2015, 2016, 2017, 2018, 2019]
dfs = []

for year in years:
    file_name = f'{year}.csv'
    df_year = pd.read_csv(file_name)
    df_year['year'] = year

    # Normalize column names BEFORE selection
    df_year.columns = df_year.columns.str.strip().str.lower().str.replace(' ', '_').str.replace(r'[().]', '', regex=True)

    rename_map = {
        'country_or_region': 'country',
        'score': 'happiness_score',
        'happinessscore': 'happiness_score',
        'economygdppercapita': 'economy_gdp_per_capita',
        'social_support': 'family',
        'healthylife_expectancy': 'health_life_expectancy',
        'healthlife_expectancy': 'health_life_expectancy',
        'freedomtomakelifechoices': 'freedom',
        'trustgovernmentcorruption': 'trust_government_corruption',
        'perceptionsofcorruption': 'trust_government_corruption',
    }
    df_year = df_year.rename(columns=rename_map)

    # Filter to required columns if present
    required_columns = ['country', 'happiness_score', 'economy_gdp_per_capita',
                        'family', 'health_life_expectancy', 'freedom',
                        'trust_government_corruption', 'generosity', 'year']

    df_year = df_year[[col for col in required_columns if col in df_year.columns]]

    dfs.append(df_year)

# Concatenate all years
df_all = pd.concat(dfs, ignore_index=True)

# Drop rows missing critical values
df_all = df_all.dropna(subset=['country', 'happiness_score'])

# Export to CSV for your Streamlit dashboard
df_all.to_csv('merged_happiness_data.csv', index=False)

# Confirm structure
print("Shape of merged dataset:", df_all.shape)
print("Available years:", sorted(df_all['year'].unique()))
df_all.head()
```

```python
# Reload merged multi-year dataset for analysis
import pandas as pd

df_all = pd.read_csv('merged_happiness_data.csv')

print("‚úÖ merged_happiness_data.csv loaded successfully.")
print(df_all.head())
```

```python
# üîπ Step 19: Advanced Insights and Predictive Modeling (Final Clean)
# ü™Ñ What to Expect:
# - Train Linear Regression on cleaned multi-year dataset
# - Visualize feature importance, predicted vs actual, residuals
# - Output evaluation metrics (MSE, R¬≤)
# ---------------------------------------------------------------

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns

# Features for prediction
features = [
    'economy_gdp_per_capita',
    'family',
    'health_life_expectancy',
    'freedom',
    'trust_government_corruption',
    'generosity'
]

# Clean and prepare data
df_clean = df_all.dropna(subset=features + ['happiness_score'])
X = df_clean[features]
y = df_clean['happiness_score']

# Train Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Feature Importance Plot (future-proof)
importance = model.coef_
plt.figure(figsize=(10, 6))
sns.barplot(x=importance, y=features, hue=features, palette='viridis', dodge=False, legend=False)
plt.title('Feature Importance on Happiness Score (Linear Regression)')
plt.xlabel('Coefficient Value')
plt.ylabel('Feature')
plt.show()

# Predicted vs Actual Plot
y_pred = model.predict(X)
plt.figure(figsize=(8, 6))
sns.scatterplot(x=y, y=y_pred, alpha=0.6)
plt.plot([y.min(), y.max()], [y.min(), y.max()], '--r')
plt.xlabel('Actual Happiness Score')
plt.ylabel('Predicted Happiness Score')
plt.title('Predicted vs Actual Happiness Scores')
plt.show()

# Residual Distribution Plot
residuals = y - y_pred
plt.figure(figsize=(8, 6))
sns.histplot(residuals, bins=30, kde=True, color='skyblue')
plt.title('Residual Distribution (Predicted - Actual)')
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.show()

# Evaluation Metrics
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)
print("‚úÖ Step 19 Complete")
print(f"Mean Squared Error: {mse:.4f}")
print(f"R^2 Score: {r2:.4f}")
```
